```python
link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[@href='/lionbeard17/followers/']")))
link.click()
# Allow some time for the followers list to load
time.sleep(5)
# Locate the followers list container using the provided class
followers_list = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='xyi19xy x1ccrb07 xtf3nb5 x1pc53ja x1lliihq x1iyjqo2 xs83m0k xz65tgg x1rife3k x1n2onr6']")))

# Initialize an empty list to store follower data
follower_data = []

# Scroll the followers list 20 times
for i in range(20):
    # Extract the text before scrolling
    followers_text_before_scroll = followers_list.text

    # Execute JavaScript to scroll down the followers list
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", followers_list)
    # Allow some time for the new followers to load
    time.sleep(3)

    # Extract the text after scrolling
    followers_text_after_scroll = followers_list.text

    # Split the follower text by newline character to get individual follower names
    followers = followers_text_after_scroll.split('\n')
    
    # Append the new follower data to the follower_data list
    follower_data.extend(followers)

    # Compare the text before and after scrolling
    print(f"Scroll {i+1}:")
    print("Followers before scroll:", followers_text_before_scroll)
    print("Followers after scroll:", followers_text_after_scroll)
    print("---------------------------------------------------")

# Create a dataframe from the follower data
follower_df = pd.DataFrame({'Follower': follower_data})

# Print the dataframe
print(follower_df)

# Save the dataframe to a CSV file
follower_df.to_csv('follower_data.csv', index=False)
```
